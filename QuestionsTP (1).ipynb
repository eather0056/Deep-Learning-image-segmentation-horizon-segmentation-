# Answer:
1. The data used to develop a neural network is typically split into three categories:
                * Training Data.
                * Validation Data.
                * and Test Data.

2. The role of each data category:

   - **Training Data**: This is the largest subset of the dataset and is used for the actual training of the neural network. The model learns to recognize patterns and make decisions based on this data. It's during this phase that the model's weights and biases are adjusted through processes like backpropagation.
   
   - **Validation Data**: This data is used to provide an unbiased evaluation of a model fit during the training phase. It helps in tuning the model's hyperparameters and provides a check against overfitting. The model doesn't learn from this data; rather, it's used to assess the model's performance and make adjustments to the learning process.
   
   - **Test Data**: After the model has been trained and validated, the test data is used to assess its performance. This data is never used in the training process and serves as a new, unseen dataset for the model. It provides a final metric to evaluate how well the model is likely to perform on real-world data.

3. The distribution of the dataset into these categories can vary based on the size of the dataset, the complexity of the model, and the specific practices of the data scientist or the requirements of the project. A common distribution is:

   - **Training Data**: Approximately 70-80% of the entire dataset.
   - **Validation Data**: Approximately 10-15% of the entire dataset.
   - **Test Data**: Approximately 10-15% of the entire dataset.

4. The differences between the three types of segmentation:

   - **Semantic Segmentation**: This type of segmentation involves labeling each pixel in the image with a class corresponding to what the pixel represents. It does not differentiate between different instances of the same class; all instances are labeled with the same class label.
   
   - **Instance Segmentation**: Similar to semantic segmentation, instance segmentation also involves labeling each pixel of the image. However, it goes one step further by distinguishing between different instances of the same class. So, if there are three cars in an image, each car would be uniquely identified.
   
   - **Panoptic Segmentation**: Panoptic segmentation combines the concepts of semantic and instance segmentation. It involves labeling each pixel with a class as in semantic segmentation, and it also differentiates between different instances of the same class, similar to instance segmentation. The term 'panoptic' refers to the unification of these two approaches, providing a comprehensive view of the scene.

5. Thsi TP lab, we are concerned with Semantic Segmentation.

6. For the segmentation problem, the labeling takes the form of an image where each pixel is assigned a label corresponding to the class of the object it belongs to. In semantic segmentation, the labels would be class-specific but not instance-specific. In instance segmentation, each object instance would have a unique label. For panoptic segmentation, the labeling would include both class and instance labels. This usually results in a mask for each object in the image, whether they are of the same class or different classes. These masks are often stored in formats that directly correspond to the image pixel data, like PNG files where each pixel's color corresponds to a class label.

1. **Main Function of the Convolution Layer**:
   The primary function of the convolution layer in a neural network is to extract features from the input image or feature map. It does this by performing a mathematical operation called convolution, which involves sliding a filter or kernel over the input and computing the dot product between the kernel and the input at each position. This process captures local dependencies in the image and generates feature maps that progressively capture higher-level features as you go deeper into the network.

3. **Weight Sharing in Convolution Layers**:
   Weight sharing refers to the use of the same filter or kernel across the entire input image. This means that instead of learning a separate set of parameters for every location in the image, the network learns one set of parameters or the filter weights that is applied to all locations. This significantly reduces the number of parameters in the network, which not only helps to prevent overfitting but also reduces computational requirements. The benefit is that the filter can detect the same feature regardless of its position in the image, which is particularly useful for image data where the same feature like an edge or a corner, can occur in different places.

4. **Stacking Convolution Layers in Deep Learning**:
   Stacking multiple convolution layers in deep learning allows the network to learn a hierarchy of features. The first layers might learn simple features like edges or textures, while deeper layers can combine these into more complex features like patterns or parts of objects, and even further layers can represent whole objects or scenes. This hierarchical feature learning is powerful for tasks like image classification, object detection, and segmentation, where understanding at different levels of abstraction is beneficial.

5. **Function of Batch Normalization Layer**:
   The batch normalization layer is used to stabilize and accelerate the training of deep neural networks. It normalizes the output of the previous activation layer by subtracting the batch mean and dividing by the batch standard deviation. Afterward, it scales and shifts the result with two learnable parameters per feature. The process of batch normalization helps to address issues like internal covariate shift where the distribution of inputs to layers changes as the parameters of the previous layers change, which can slow down training and make it harder to train models with higher learning rates.

6. **Point of Layer Normalization Layer**:
   Layer normalization is similar to batch normalization but instead of normalizing across the batch, it normalizes across the features. In layer normalization, the mean and variance are computed across all the features for a single data-point. It is particularly useful for recurrent neural networks (RNNs) and is not dependent on the batch size, making it suitable for tasks with a dynamic batch size.

7. **Use of Batch Normalization Over Layer Normalization**:
   Batch normalization tends to work better than layer normalization in convolutional neural networks dealing with image data because it normalizes the feature maps across the batch which is effective in stabilizing the mean and variance of each layer's outputs. Layer normalization might be more useful in recurrent neural networks or when the batch size is small or varies significantly.

8. **Types of Activation Functions**:
   - ReLU (Rectified Linear Unit)
   - Sigmoid
   - Tanh (Hyperbolic Tangent)
   - Leaky ReLU
   - Parametric ReLU
   - Exponential Linear Unit (ELU)
   - Softmax (often used in the output layer for classification)

9. **Purpose of Activation Functions**:
   Activation functions introduce non-linear properties to the network which enables the network to learn more complex functions or representations from the data. Without non-linearity, no matter how many layers the network has, it would still behave like a single-layer network because the composition of linear functions is still a linear function. Activation functions allow networks to approximate non-linear functions, making them capable of performing tasks like classification and regression on complex datasets.

# Answer:

1. In the binary cross-entropy formula:
   - N is the number of observations or samples we're looking at.
   - $ y_i $ represents the actual label of the $ i^{th}$ observation, which can be either 0 or 1 in binary classification.
   - $ p(y_i) $ is the predicted probability that the $ i^{th} $observation is 1 according to the model.


2. The term $ -y_i log(p(y_i))$ equals:
   - **2.1.** If the target is 1 and the prediction is also 1, the term becomes $ -1 \log(1) $ which is $ 0 $ because the log of 1 is 0.
   - **2.2.** If the target is 1 and the prediction tends to 0, the term approaches infinity, as the log of a number very close to 0 is a very large negative number, and we're taking the negative of that.

3. The term $ -y_i log(p(y_i)) $ represents the loss for an individual observation when the actual value is 1. It measures how far off the prediction is from the actual value. If the prediction is perfect, the loss is 0; the worse the prediction, the higher the loss.

4. The term $ -(1-y_i) log(1-p(y_i)) $ equals:
   - **4.1.** If the target is 0 and the prediction is 1, it turns into $ -0 \log(0) $, which is $ 0 $ because $ 0 $ times anything is $ 0 $.
   - **4.2.** If the target is 0 and the prediction tends to 0, the term becomes $ -0 \log(1) $, which is $ 0 $, again because $ 0 $ times anything is $ 0 $.

5. The term $ -(1-y_i) log(1-p(y_i)) $ represents the loss for an individual observation when the actual value is 0. It captures how off the mark the model's prediction is when it's supposed to predict a 0.

6. Binary cross-entropy is supposed to measure the "distance" between the true labels and the predicted probabilities. A perfect model would have a cross-entropy of 0, while any deviation from the true labels increases the cross-entropy.

7. `BCEWithLogitsLoss` combines a sigmoid layer with the binary cross-entropy loss in one single class. This is more numerically stable than using a plain Sigmoid followed by a BCELoss separately because it handles the log and exp operations in a way that avoids underflow or overflow problems. So, you should use `BCEWithLogitsLoss` when you haven't applied a sigmoid activation function to your outputs.

8. The drawbacks of binary cross-entropy include:
   - It assumes that each observation's label is independent of all other observations, which might not be the case.
   - For imbalanced datasets, where the number of observations in one class significantly outnumbers the other, binary cross-entropy can cause the model to favor the majority class.

9. Other loss functions suitable for image segmentation:
   - Dice Loss: Useful when dealing with imbalanced datasets, as it's a measure of overlap between two samples.
   - Jaccard/Intersection over Union (IoU) Loss: Similar to Dice Loss, it measures the size of the intersection divided by the size of the union of two sets of samples.
   - Focal Loss: Designed to address class imbalance by focusing more on hard-to-classify examples.
   - Tversky Loss: Generalizes Dice Loss by adding weights to false positives and false negatives.

1. **Training vs. Inference Phase**:
   - The **training phase** is when the neural network learns from a dataset. It adjusts its weights through a process of optimization, trying to minimize the loss function, which measures the error between the model's predictions and the actual data.
   - The **inference phase** occurs after the model is trained. It's when you use the model to make predictions on new, unseen data. The model's weights are no longer being updated; you're simply applying the learned model to get outputs from inputs.

2. **Goal of Gradient Descent**:
   The goal of gradient descent is to minimize the loss function. It does this by calculating the gradient of the loss function with respect to the model's weights and then changing the weights in the opposite direction of the gradient to reduce the loss.

3. **Goal of Back Propagation**:
   Back propagation is the method used to calculate the gradient of the loss function with respect to the model's weights. It efficiently computes the gradient by propagating the error backward through the network layers, hence the name.

4. **New Weights Calculation**:
   The relation between the new weight $ w_{\text{new}} $, the previous weights $ w $, and the cost function $ Q $ is given by the update rule:
   $w_{new}$ = w - $ \eta $ $\nabla$ Q(w)
   Here, $ \eta $ represents the learning rate, and $ \nabla Q(w) $ represents the gradient of the cost function with respect to the weights.

5. **Gradient Descent on Simple Function**:
   If we want to apply gradient descent to the function $ f(x) = 3x^2 + 5x + 4 $. To do this, we will first calculate the derivative of $ f $ with respect to $ x $, which gives the gradient. Then, we will pick a starting value for $ x $ and update it iteratively using the gradient descent update rule.

6. **Stochastic Gradient Descent (SGD)**:
   SGD is called stochastic because, instead of computing the gradient based on the entire dataset (as in standard gradient descent), it uses a random subset of the data at each step. This randomness helps to speed up the computations and can also help to avoid local minima.

7. **Batch Size**:
   The batch size is the number of training samples used in one iteration of model training. In one batch, the model's weights are updated once after seeing all the samples in that batch.

8. **Limitation of Batch Size**:
   Batch size is limited in practice due to hardware constraints. Larger batches require more memory and computational power. Additionally, very large batches may lead to poorer convergence properties during training.

9. **Epochs Number**:
   An epoch is one complete pass through the entire training dataset. If we have 1000 samples and we want to use a batch size of 100, it would take 10 batches to complete one epoch.

10. **Steps per Epoch**:
   Steps per epoch is the number of batches of data the model will process before it considers that one epoch is completed. It is the total number of training samples divided by the batch size.

11. **Relation Between Dataset Size, Batch Size, and Steps per Epoch**: The relation is given by:
    
   $ \text{Steps per epoch} = \frac{\text{Size of training
   dataset}}{\text{Batch size}} $

   This tells you how many steps (batches) you need to cover the entire training dataset in one epoch.

12. **Effect of Learning Rate**: The learning rate controls how much the weights are updated during training. A very small learning rate may lead to very slow convergence, while a very large learning rate can cause the training to oscillate or even diverge.

13. **Effect of Batch Size**: A smaller batch size usually provides a regularizing effect and lower generalization error. However, it can also mean that the optimization is noisier and may take longer to converge. A larger batch size provides a more accurate estimate of the gradient but might lead to overfitting.

14. **Too Many Epochs**: If the number of epochs is too large, the model may overfit to the training data. This means it's learning the noise and the details in the training data to the extent that it negatively impacts the performance on new, unseen data.
